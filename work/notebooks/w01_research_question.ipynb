## Section 2 — The Question: Decision, Action, Cost of a Wrong Call

**The Decision:** Reviewers must select a prioritized subset of pages from
hundreds of potential candidates to maximize the impact of their limited
weekly review hours.

**The Action:** For each high-priority page, reviewers take a specific action
such as updating stale content, expanding thin pages, or setting up monitoring.

**The Cost of a False Alarm:** Flagging a healthy page mistakenly wastes
valuable analyst time. This can drain a full week of productivity across
multiple errors while actual issues go unnoticed.

**The Cost of a Missed Page:** Failing to identify a declining page allows
traffic and impressions to drop silently. This leads to a much deeper and
harder to recover ranking loss over time.
